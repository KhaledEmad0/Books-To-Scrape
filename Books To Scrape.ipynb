```python
ratings = {
    'One': 1,
    'Two':2,
    'Three':3,
    'Four':4,
    'Five':5
}
```

```python
http = 'https://books.toscrape.com/'

with open('Books To Scrape.csv', 'w', encoding='utf-8') as f:
    writer = csv.writer(f)
    writer.writerow(['title', 'rate', 'category', 'price'])
    
    for link in urls:
        href = link.get('href')
        url = http+href
        response = requests.get(url)
        soup = BeautifulSoup(response.text, 'html.parser')
        books = soup.find('ol', attrs={'class': 'row'}).find_all('li')
        print(url)
        for book in books:
            title = book.find('article').find('h3').find('a').get('title').strip()
            rate = ratings[book.find('article').find('p').get('class')[1]]
            category = link.text.strip()
            price = float(book.find('article').find_all('div')[-1].find('p', attrs={'class':'price_color'}).get_text()[2:])
            writer.writerow([title, rate, category, price])
```
